```python
import pandas as pd
from pandas import DataFrame
import numpy as np
from statistics import mean
import statistics

nd = pd.read_csv('/home/kusumapani/Desktop/empolis/dataset/nelson_rules_time_series.csv')
a = list(nd['value'])
time = nd[['timestamp']].to_numpy()
```

```python
def R1(l):
    avg = mean(a[0:100])
    std = statistics.stdev(a[0:100])
    upper = avg + (std * 3)
    lower = avg - (std * 3)
    final = ["1" if i<= lower else "1" if i>= upper else "0" for i in l]
    return final
    
```

```python
def R2(l):
    res = [[l[0]]]
    avg = mean(l[0:100])
    
    for i in range(1, len(l)):
        if ((l[i-1] >= avg) and (l[i] >= avg)):
            res[-1].append(l[i])            
        elif ((l[i-1] <= avg) and (l[i] <= avg)):
            res[-1].append(l[i])
        else:
            res.append([l[i]])
    #print(res)
    final = ["1" if len(sublist) >= 9 else "0" for sublist in res for val in sublist]
    return final
```

```python
def R3(l):
    res = [[l[0],l[1]]]
    i = 2
    while (i < len(l)):         
        if (l[i-2] < l[i-1]) and (l[i-1] < l[i]):
            res[-1].append(l[i])
        elif (l[i-2] > l[i-1]) and (l[i-1] > l[i]):
            res[-1].append(l[i])            
        else:
            res.append([l[i]])
            i+=1
            if ((i) < len(l)):
                res[-1].append(l[i])        
        i+=1
    print(res)
    final = ["1" if len(sublist) >= 6 else "0" for sublist in res for val in sublist]
    return final
```

```python
def R4(l):
    res = [[l[0],l[1]]]
    i = 2
    while (i < len(l)):         
        if (l[i-2] < l[i-1]) and (l[i-1] > l[i]):
            res[-1].append(l[i])
        elif (l[i-2] > l[i-1]) and (l[i-1] < l[i]):
            res[-1].append(l[i])            
        else:
            res.append([l[i]])
            i+=1
            if ((i) < len(l)):
                res[-1].append(l[i])        
        i+=1
    final = ["1" if len(sublist) >= 14 else "0" for sublist in res for val in sublist]
    return final
```

```python
file=open("/home/kusumapani/Desktop/empolis/Soultions_CSV/new_nel.csv", 'w')
R11 = R1(a)
R22 = R2(a)
R33 = R3(a)
R44 = R4(a)
for i in range(len(a)):
    
    output=str(int(time[i]))
    result = ""
    if (R11[i]) != "0":
        result+= "1;"
    
    
    if (R22[i]) != "0":
        result+= "2;"
        
    
    if (R33[i]) != "0":
        result+= "3;"
        
    
    if (R44[i]) != "0":
        result+= "4;"
        
    output+=","+result[:-1]+"\n"
    file.write(output)
    
file.close()
```
